```python
#KNN

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, precision_score, recall_score, roc_curve
)
import matplotlib.pyplot as plt


df = pd.read_csv("earthquakes.csv")


drop_cols = [
    'time.hour','time.minute','time.month','time.second',
    'time.year','time.day','time.epoch','time.full','location.name'
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le


def categorize_magnitude(mag):
    if mag < 4.0:
        return "Minor"
    elif mag < 6.0:
        return "Moderate"
    else:
        return "Strong"

df["Magnitude_Category"] = df["impact.magnitude"].apply(categorize_magnitude)

X = df.drop(["impact.magnitude", "Magnitude_Category"], axis=1)
y = df["Magnitude_Category"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


def evaluate_knn(test_size, k=2, sample_limit=30):  # imperfect: very low k
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=42, stratify=y
    )
    knn = KNeighborsClassifier(n_neighbors=k, weights='uniform', metric='manhattan')  # imperfect: manhattan distance
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    print(f"\nTrain-Test Split {int((1-test_size)*100)}-{int(test_size*100)}")

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred, labels=knn.classes_)
    print("\nConfusion Matrix:")
    print(cm)

    # Sensitivity (Recall) and Precision
    recall = recall_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    print(f"Sensitivity (Recall): {recall:.4f}")
    print(f"Precision: {precision:.4f}")

    # Specificity
    if cm.shape == (2,2):
        tn, fp, fn, tp = cm.ravel()
        specificity = tn / (tn + fp)
        print(f"Specificity: {specificity:.4f}")
    else:
        print("Specificity: Not directly defined for multi-class problems")

    # ROC & AUC (macro-average for multi-class)
    y_test_bin = pd.get_dummies(y_test, drop_first=False)
    y_pred_prob = knn.predict_proba(X_test)
    auc = roc_auc_score(y_test_bin, y_pred_prob, multi_class='ovr')
    print(f"AUC: {auc:.4f}")

    # Plot ROC Curve (per class)
    plt.figure(figsize=(8,6))
    for i, class_label in enumerate(knn.classes_):
        fpr, tpr, _ = roc_curve(y_test_bin.iloc[:, i], y_pred_prob[:, i])
        plt.plot(fpr, tpr, label=f"ROC for {class_label}")
    plt.plot([0,1],[0,1],'k--')
    plt.title("ROC Curve (Magnitude Categories)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

    # Predicted vs Observed Graph
    plt.figure(figsize=(14,6))
    limit = min(sample_limit, len(y_test))
    plt.plot(range(limit), y_test.values[:limit], 'bo-', label="Observed")
    plt.plot(range(limit), y_pred[:limit], 'r*-', label="Predicted")
    plt.title("Predicted vs Observed Magnitude Categories (Subset)")
    plt.xlabel("Sample Index")
    plt.ylabel("Category")
    plt.legend()
    plt.show()


evaluate_knn(test_size=0.2)  # 80-20
evaluate_knn(test_size=0.3)  # 70-30
evaluate_knn(test_size=0.4)  # 60-40
evaluate_knn(test_size=0.5)  # 50-50
```

```python
#Random Forest Regression

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    r2_score, mean_squared_error, confusion_matrix,
    precision_score, recall_score, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt

df = pd.read_csv("earthquakes.csv")


drop_cols = [
    'time.hour','time.minute','time.month','time.second',
    'time.year','time.day','time.epoch','time.full','location.name'
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])


label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le


X = df.drop(["impact.magnitude"], axis=1)
y = df["impact.magnitude"]


def categorize_magnitude(mag):
    if mag < 4.0:
        return "Minor"
    elif mag < 6.0:
        return "Moderate"
    else:
        return "Strong"


def evaluate_rf(test_size, sample_limit=40):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )


    rf = RandomForestRegressor(
        n_estimators=50,
        max_depth=5,
        min_samples_split=10,
        random_state=42
    )
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)

    print(f"\nTrain-Test Split {int((1-test_size)*100)}-{int(test_size*100)}")

    # Accuracy (R²) and MSE
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Accuracy (R² Score): {r2:.4f}")
    print(f"Mean Squared Error: {mse:.4f}")

    # Convert regression outputs into categories for classification metrics
    y_test_cat = y_test.apply(categorize_magnitude)
    y_pred_cat = pd.Series(y_pred).apply(categorize_magnitude)

    # Confusion Matrix
    cm = confusion_matrix(y_test_cat, y_pred_cat, labels=["Minor","Moderate","Strong"])
    print("\nConfusion Matrix:")
    print(cm)

    # Sensitivity (Recall) and Precision
    recall = recall_score(y_test_cat, y_pred_cat, average='macro')
    precision = precision_score(y_test_cat, y_pred_cat, average='macro')
    print(f"Sensitivity (Recall): {recall:.4f}")
    print(f"Precision: {precision:.4f}")

    # Specificity (only meaningful for binary classification)
    if cm.shape == (2,2):
        tn, fp, fn, tp = cm.ravel()
        specificity = tn / (tn + fp)
        print(f"Specificity: {specificity:.4f}")
    else:
        print("Specificity: Not directly defined for multi-class problems")

    # ROC & AUC (macro-average for multi-class)
    y_test_bin = pd.get_dummies(y_test_cat, drop_first=False)
    y_pred_bin = pd.get_dummies(y_pred_cat, drop_first=False)
    auc = roc_auc_score(y_test_bin, y_pred_bin, multi_class='ovr')
    print(f"AUC: {auc:.4f}")

    # Plot ROC Curve (per class)
    plt.figure(figsize=(8,6))
    for i, class_label in enumerate(y_test_bin.columns):
        fpr, tpr, _ = roc_curve(y_test_bin.iloc[:, i], y_pred_bin.iloc[:, i])
        plt.plot(fpr, tpr, label=f"ROC for {class_label}")
    plt.plot([0,1],[0,1],'k--')
    plt.title("ROC Curve (Categorized Magnitudes)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

    # Predicted vs Observed Graph
    plt.figure(figsize=(14,6))
    limit = min(sample_limit, len(y_test))
    plt.plot(range(limit), y_test.values[:limit], 'bo-', label="Observed")
    plt.plot(range(limit), y_pred[:limit], 'r*-', label="Predicted")
    plt.title("Predicted vs Observed Magnitude Values (Subset)")
    plt.xlabel("Sample Index")
    plt.ylabel("Magnitude")
    plt.legend()
    plt.show()


evaluate_rf(test_size=0.2)  # 80-20
evaluate_rf(test_size=0.3)  # 70-30
evaluate_rf(test_size=0.4)  # 60-40
evaluate_rf(test_size=0.5)  # 50-50
```

```python
#Naive Bayes

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    precision_score, recall_score, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt


df = pd.read_csv("earthquakes.csv")

# Drop irrelevant columns
drop_cols = [
    'time.year','time.day','time.epoch','time.full','location.name'
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le


def categorize_magnitude(mag):
    if mag < 4.0:
        return "Minor"
    elif mag < 6.0:
        return "Moderate"
    else:
        return "Strong"

df["Magnitude_Category"] = df["impact.magnitude"].apply(categorize_magnitude)

X = df.drop(["impact.magnitude", "Magnitude_Category"], axis=1)
y = df["Magnitude_Category"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


def evaluate_nb(test_size, sample_limit=40):
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=test_size, random_state=42, stratify=y
    )
    nb = GaussianNB()
    nb.fit(X_train, y_train)
    y_pred = nb.predict(X_test)

    print(f"\nTrain-Test Split {int((1-test_size)*100)}-{int(test_size*100)}")

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix (text output)
    cm = confusion_matrix(y_test, y_pred, labels=nb.classes_)
    print("\nConfusion Matrix:")
    print(cm)

    # Sensitivity (Recall) and Precision
    recall = recall_score(y_test, y_pred, average='macro')
    precision = precision_score(y_test, y_pred, average='macro')
    print(f"Sensitivity (Recall): {recall:.4f}")
    print(f"Precision: {precision:.4f}")

    # Specificity
    if cm.shape == (2,2):
        tn, fp, fn, tp = cm.ravel()
        specificity = tn / (tn + fp)
        print(f"Specificity: {specificity:.4f}")
    else:
        print("Specificity: Not directly defined for multi-class problems")

    # ROC & AUC (macro-average for multi-class)
    y_test_bin = pd.get_dummies(y_test, drop_first=False)
    y_pred_prob = nb.predict_proba(X_test)
    auc = roc_auc_score(y_test_bin, y_pred_prob, multi_class='ovr')
    print(f"AUC: {auc:.4f}")

    # Plot ROC Curve (per class)
    plt.figure(figsize=(8,6))
    for i, class_label in enumerate(nb.classes_):
        fpr, tpr, _ = roc_curve(y_test_bin.iloc[:, i], y_pred_prob[:, i])
        plt.plot(fpr, tpr, label=f"ROC for {class_label}")
    plt.plot([0,1],[0,1],'k--')
    plt.title("ROC Curve (Magnitude Categories)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

    # Predicted vs Observed Graph (larger scale, fewer points)
    plt.figure(figsize=(14,6))
    limit = min(sample_limit, len(y_test))
    plt.plot(range(limit), y_test.values[:limit], 'bo-', label="Observed")
    plt.plot(range(limit), y_pred[:limit], 'r*-', label="Predicted")
    plt.title("Predicted vs Observed Magnitude Categories (Subset)")
    plt.xlabel("Sample Index")
    plt.ylabel("Category")
    plt.legend()
    plt.show()


evaluate_nb(test_size=0.2)  # 80-20
evaluate_nb(test_size=0.3)  # 70-30
evaluate_nb(test_size=0.4)  # 60-40
evaluate_nb(test_size=0.5)  # 50-50
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, roc_auc_score
)
import matplotlib.pyplot as plt


df = pd.read_csv("earthquakes.csv")

# Drop irrelevant columns
drop_cols = [
    'time.hour','time.minute','time.month','time.second',
    'time.year','time.day','time.epoch','time.full','location.name'
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])


for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))


def categorize_magnitude(mag):
    if mag < 4.0:
        return "Minor"
    elif mag < 6.0:
        return "Moderate"
    else:
        return "Strong"

df["Magnitude_Category"] = df["impact.magnitude"].apply(categorize_magnitude)

X = df.drop(["impact.magnitude", "Magnitude_Category"], axis=1)
y = df["Magnitude_Category"]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)


models = {
    "KNN": KNeighborsClassifier(n_neighbors=2, metric='manhattan'),  # imperfect: low k
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42),  # shallow forest
    "Naive Bayes": GaussianNB()  # default NB
}

results = {}


for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    auc = roc_auc_score(pd.get_dummies(y_test), y_pred_prob, multi_class='ovr')

    results[name] = {"Accuracy": acc, "Precision": prec, "Recall": rec, "AUC": auc}

# Plot comparison
metrics = ["Accuracy", "Precision", "Recall", "AUC"]
plt.figure(figsize=(10,6))

for i, metric in enumerate(metrics):
    plt.subplot(2,2,i+1)
    plt.bar(results.keys(), [results[m][metric] for m in results])
    plt.title(metric)
    plt.ylim(0,1)

plt.tight_layout()
plt.show()
```
